```python
df.iloc[2:12,::] # from 2nd row to last row and all the columns selected
```

```python
data = ('col1,col2,col3\n'
       'x,y,1\n'
       'a,b,2\n'
       'd,c,3\n')
type(data)
```

```python
pd.read_csv(StringIO(data)) # another way to read csv data 
```

```python
data = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11')
print(data)
```

```python
df = pd.read_csv(StringIO(data),dtype =object)
df
```

```python
df['a']
```

```python
# making all the column as different data type
df = pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'int64'})
df
```

```python
df['a'][1] # serias a 2nd value
```

```python
df.dtypes  # types of each features
```
